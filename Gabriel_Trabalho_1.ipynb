```
#Insira seu código abaixo
def resizeImage(img1, img2):
  width = int(img2.shape[1])
  height = int(img2.shape[0]) 
  img1 = cv2.resize(img1, (width, height))

  return img1


def addBlendingEffect (firstImage, secondImage, weight):
  firstImageGray = cv2.cvtColor(firstImage, cv2.COLOR_BGR2GRAY)
  secondImageGray = cv2.cvtColor(secondImage, cv2.COLOR_BGR2GRAY)

  firstImage = resizeImage(firstImage, secondImage)
  firstImageGray = resizeImage(firstImageGray, secondImageGray)

  mask = firstImageGray - secondImageGray
  ret, mask = cv2.threshold(mask, 0, 255, cv2.THRESH_BINARY)

  copyImg = firstImage.copy()
  altura, largura = mask.shape

  for y in range(0, altura):
    for x in range(0, largura):
      if mask.item(y, x) > 0:
        finalPixelBlue = firstImage.item(y, x, 0) * (1.0 - weight) + secondImage.item(y, x, 0) * weight
        finalPixelGreen = firstImage.item(y, x, 1) * (1.0 - weight) + secondImage.item(y, x, 1) * weight
        finalPixelRed = firstImage.item(y, x, 2) * (1.0 - weight) + secondImage.item(y, x, 2) * weight

        copyImg.itemset((y, x, 0), finalPixelBlue)
        copyImg.itemset((y, x, 1), finalPixelGreen)
        copyImg.itemset((y, x, 2), finalPixelRed)

  return copyImg



def memeGeneratorWithBlending(imagem1, fundo):
    finalImage = addBlendingEffect(imagem1, fundo, 0.1)
    
    plt.imshow(finalImage)


original = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/plantaçao.png")
marca = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/selo.png")


original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
marca = cv2.cvtColor(marca, cv2.COLOR_BGR2RGB)

plt.imshow(original)
plt.figure()
plt.imshow(marca)
plt.figure()


memeGeneratorWithBlending(original,marca)
```

```
#Insira seu código abaixo
#from __future__ import print_function
#from __future__ import division
import argparse



src_base = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/gabriel.png")
src_base = cv2.cvtColor(src_base, cv2.COLOR_BGR2RGB)

src_test1 = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/Lenna.png")
src_test1 = cv2.cvtColor(src_test1, cv2.COLOR_BGR2RGB)

src_test2 = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/brilhoecontraste.jpg")
src_test2 = cv2.cvtColor(src_test2, cv2.COLOR_BGR2RGB)

src_test3 = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/bob.png")
src_test3 = cv2.cvtColor(src_test3, cv2.COLOR_BGR2RGB)

src_test4 = cv2.imread("/content/drive/MyDrive/ProcessamentoImagem/gabriel2.png")
src_test4 = cv2.cvtColor(src_test4, cv2.COLOR_BGR2RGB)


#src_base = cv2.imread(input1)
#src_test1 = cv2.imread(input2)
#src_test2 = cv2.imread(input3)
#src_test3 = cv2.imread(input4)
#src_test4 = cv2.imread(input5)

if src_base is None or src_test1 is None or src_test2 is None or src_test3 is None or src_test4 is None:
    print('Could not open or find the images!')
    exit(0)
hsv_base = cv2.cvtColor(src_base, cv2.COLOR_BGR2HSV)
hsv_test1 = cv2.cvtColor(src_test1, cv2.COLOR_BGR2HSV)
hsv_test2 = cv2.cvtColor(src_test2, cv2.COLOR_BGR2HSV)
hsv_test3 = cv2.cvtColor(src_test3, cv2.COLOR_BGR2HSV)
hsv_test4 = cv2.cvtColor(src_test4, cv2.COLOR_BGR2HSV)

hsv_half_down = hsv_base[hsv_base.shape[0]//2:,:]

h_bins = 50
s_bins = 60
histSize = [h_bins, s_bins]
# hue varies from 0 to 179, saturation from 0 to 255
h_ranges = [0, 180]
s_ranges = [0, 256]
ranges = h_ranges + s_ranges # concat lists
# Use the 0-th and 1-st channels
channels = [0, 1]

hist_base = cv2.calcHist([hsv_base], channels, None, histSize, ranges, accumulate=False)
cv2.normalize(hist_base, hist_base, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

hist_half_down = cv2.calcHist([hsv_half_down], channels, None, histSize, ranges, accumulate=False)
cv2.normalize(hist_half_down, hist_half_down, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

hist_test1 = cv2.calcHist([hsv_test1], channels, None, histSize, ranges, accumulate=False)
cv2.normalize(hist_test1, hist_test1, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

hist_test2 = cv2.calcHist([hsv_test2], channels, None, histSize, ranges, accumulate=False)
cv2.normalize(hist_test2, hist_test2, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

hist_test3 = cv2.calcHist([hsv_test3], channels, None, histSize, ranges, accumulate=False)
cv2.normalize(hist_test3, hist_test3, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

hist_test4 = cv2.calcHist([hsv_test4], channels, None, histSize, ranges, accumulate=False)
cv2.normalize(hist_test4, hist_test4, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

for compare_method in range(4):
    base_base = cv2.compareHist(hist_base, hist_base, compare_method)
    base_half = cv2.compareHist(hist_base, hist_half_down, compare_method)
    base_test1 = cv2.compareHist(hist_base, hist_test1, compare_method)
    base_test2 = cv2.compareHist(hist_base, hist_test2, compare_method)
    base_test3 = cv2.compareHist(hist_base, hist_test3, compare_method)
    base_test4 = cv2.compareHist(hist_base, hist_test4, compare_method)

    print('Method:', compare_method, 'Perfect, Base-Half, Base-Test(1), Base-Test(2), Base-Test(3), Base-Test(4) :',\
          base_base, '/', base_half, '/', base_test1, '/', base_test2, '/', base_test3, '/', base_test4)
    
    print("A FOTO MAIS PARECIDA É A FOTO DOIS")
    
```
